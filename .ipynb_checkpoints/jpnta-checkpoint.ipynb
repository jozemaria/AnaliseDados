```python
# Importando as bibliotecas

import pandas as pd
%matplotlib inline
```

```python
#Criando o DataFrame e pulando as duas primeiras linhas e as ultimas 16 linhas

df = pd.read_excel('2121_0.xlsx', skiprows=2, skipfooter=16)
df.head()
```

```python
# Vendo as 5 ultimas linhas

df.tail(20)
```

```python
# Verificando o tipos de dados

df.dtypes
```

```python
# Criando um novo DataFrame sem os valores ND em Público acumulado
df_novo = df[df['Público acumulado'] != 'ND']

# Vendo os tipos dos dados
df_novo.dtypes
```

```python
# Convertendo os dados par inteiros

df_novo['Público acumulado'] = pd.to_numeric(df_novo['Público acumulado'])

# Vendo os tipos dos dados
df_novo.dtypes
```

```python
# Quais anos tiveram mais filmes lançados

df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
# Plotar um grafico com os lacamentos por ano
df_novo['Data de lançamento'].dt.year.value_counts().plot.bar()
```

```python
# Qual filme teve maior bilheteria 
df_novo[df_novo['Renda (R$) acumulada'] == df_novo['Renda (R$) acumulada'].max()]
```
